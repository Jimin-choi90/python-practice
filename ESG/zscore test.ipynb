```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df = pd.read_csv('ESG_Growth_Analysis.csv')
df.info()
```

```python
df['Periods'] = pd.to_datetime(df['Periods'], format='%m/%d/%Y')
df = df.sort_values(by=['Periods'], ascending=False)
```

```python
df_z = df

df_z['MSCI E mean'] = df_z.groupby(['Periods'])['MSCI E'].transform('mean')
df_z['MSCI E std'] = df_z.groupby(['Periods'])['MSCI E'].transform('std')
df_z['MSCI E z'] = (df_z['MSCI E'] - df_z['MSCI E mean']) / df_z['MSCI E std']
df_z['MSCI E z pct'] = df_z.groupby(['Periods'])['MSCI E z'].rank(ascending=False, pct=True)

df_z
```

```python
from scipy.stats import zscore

df_z2 = df[['Unnamed: 0', 'Company Name', 'Periods', 'Ticker', 'Region', 'Country', 'Sector', 'Industry', 'MSCI Rating']].join(df.dropna().groupby(['Periods']).transform(zscore))
df_z2
```

```python
# 어떻게 Ticker와 Periods 칼럼은 유지할 수 있을까?

df_z3 = df.dropna().groupby(['Periods']).transform(zscore)
df_z3 = df[['MSCI E', 'MSCI S', 'MSCI G', 'EPS', 'Net Income', 'Asset']].add_suffix("_z")
df_z3
```

```python

df = pd.merge(df, df_z3, on=[['Ticker', 'Periods']], how='left')
```
