```python
import pandas as pd
```

```python
df = pd.read_csv('CS_TS_Sample.csv')
```

```python
df = df.sort_values(by=['Company Name', 'Periods'])
```

```python
df['shifted_return'] = df['Universe Returns'].shift(1)
```

```python
df
```

```python
import statsmodels.api as sm
```

```python
df = df.dropna(subset=['shifted_return', 'F1', 'F2', 'F3'])
```

```python
y = df['shifted_return']
X = df[['F1', 'F2', 'F3']]
X = sm.add_constant(X)
```

```python
model = sm.OLS(y, X).fit()
```

```python
model.summary()
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
X = df[['F1', 'F2', 'F3']]
```

```python
pc_comps = pca.fit_transform(X)
```

```python
pca_df = pd.DataFrame(pc_comps, columns=['PC1', 'PC2'])
```

```python
pca.explained_variance_ratio_
```

```python
pca_df
```

```python
X = pca_df.values
```

```python
X = sm.add_constant(X)
```
