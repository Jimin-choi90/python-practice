```python
# F2, F3, F4의 Z-score 계산
```

```python
import pandas as pd
```

```python
df = pd.read_csv('Cross-Sectional Analysis - Sample data.csv')
df
```

```python
df.info()
```

```python
df['F4']
```

```python
# F4를 object에서 float으로 전환

def process_number(data):
    return float(data.replace(',',''))

df['F4'] = df['F4'].astype('str')
df['F4'] = df['F4'].apply(process_number)
```

```python
df.info()
```

```python
df
```

```python
# F2, F3, F4 Z-score 계산

df[['F2_Z', 'F3_Z', 'F4_Z']] = (df[['F2', 'F3', 'F4']] - df[['F2', 'F3', 'F4']].mean()) / df[['F2', 'F3', 'F4']].std()
df
```

```python
# F2, F3, F4의 국가별 mean 계산

result = (
    df
    .groupby(['Country'])
    [['F2', 'F3', 'F4']].mean()
)
```

```python
# 칼럼헤드 변경

result = result.rename({
    'F2':'F2_ctr_mean',
    'F3':'F3_ctr_mean',
    'F4':'F4_ctr_mean'
}, axis=1
)
```

```python
result[:5]
```

```python
# 합병

merged_df = pd.merge(df, result, on = ['Country'], how = 'left')
merged_df
```

```python
# F2, F3, F4의 국가별 표준편차 계산

result_std = (
    df
    .groupby(['Country'])
    [['F2', 'F3', 'F4']].std()
)
```

```python
# 인덱스 이름 변경

result_std = result_std.rename({
    'F2':'F2_ctr_std',
    'F3':'F3_ctr_std',
    'F4':'F4_ctr_std'
}, axis=1)
```

```python
result_std[:10]
```

```python
# 합병

merged_df = pd.merge(merged_df, result_std, on = ['Country'], how = 'left')
merged_df
```

```python
# Q.이렇게 하는것보다 result로 하면 이점? 이걸 그대로 Z 계산할때 쓰면 단점?
# 이렇게 하면 데이터프레임이 저장되지 않음. result 방식으로 해야 저장 됨.

merged_df.groupby(['Country'])[['F2', 'F3', 'F4']].mean()[:5]
```

```python
merged_df
```

```python
# 국가별 Z-score 계산 시, Columns must be same length as key 에러 (F2 하나씩 하면 정상작동)
# 에러메세지의 length는 # of columns가 아니라, 각 칼럼의 # of rows이다. 맞춰주고 난 뒤 시작해야 함.

merged_df['F2_ctr_Z'] = (merged_df['F2'] - merged_df['F2_ctr_mean']) / merged_df['F2_ctr_std']
merged_df['F3_ctr_Z'] = (merged_df['F3'] - merged_df['F3_ctr_mean']) / merged_df['F3_ctr_std']
merged_df['F4_ctr_Z'] = (merged_df['F4'] - merged_df['F4_ctr_mean']) / merged_df['F4_ctr_std']
merged_df
```

```python
# Q.아래 방식으로는 왜 stat이 NA가 나오는지?
# F2_sct_mean, F2_sct_std 칼럼을 각각 따로 만들어서 해야하나봄.
# df[['F2_Z', 'F3_Z', 'F4_Z']] = (df[['F2', 'F3', 'F4']] - df[['F2', 'F3', 'F4']].mean()) / df[['F2', 'F3', 'F4']].std()
  # 이게 작동했던 이유는 groupby가 없어서인듯.

merged_df['F2_sct_mean'] = merged_df.groupby(['Sector'])['F2'].mean()
merged_df['F2_sct_Z'] = (merged_df['F2'] - merged_df.groupby(['Sector'])['F2'].mean()) / merged_df.groupby(['Sector'])['F2'].std()
merged_df
```

```python
# F2_sct_mean & _std

result_mean = (
    merged_df
    .groupby(['Sector'])
    ['F2'].mean()
    .to_frame()
)

result_mean = result_mean.rename({'F2':'F2_sct_mean'}, axis=1)

result_std = (
    merged_df
    .groupby(['Sector'])
    ['F2'].std()
    .to_frame()
)

result_std = result_std.rename({'F2':'F2_sct_std'}, axis=1)
```

```python
merged_df.drop(['F2_sct_mean_x'], axis=1)
```

```python
merged_df = pd.merge(merged_df, result_mean, on=['Sector'], how='left')
merged_df = pd.merge(merged_df, result_std, on=['Sector'], how='left')
merged_df
```

```python
merged_df['F2_sct_Z'] = (merged_df['F2'] - merged_df['F2_sct_mean']) / merged_df['F2_sct_std']
merged_df
```

---------------

```python
# F2, F3, F4 Z-score의 단순합산

merged_df['Z_sum'] = merged_df['F2_Z'] + merged_df['F3_Z'] + merged_df['F4_Z']
merged_df
```

```python
pd.set_option('display.max_columns',None)
```

```python
merged_df
```

```python
merged_df['Sector'].unique()
```

```python
# Z_sum의 합 중 NA는 섹터평균으로 치환

merged_df['Z_sum_NA_sct_mean'] = merged_df.groupby(['Sector'])['Z_sum'].transform(lambda x: x.fillna(x.mean()))
merged_df
```

```python
merged_df.to_csv('ncheck.csv')
```

```python
# F1 float으로 변환

merged_df.info()
```

```python
merged_df['F1'] = merged_df['F1'].astype('str')
merged_df['F1'] = merged_df['F1'].apply(process_number)
```

```python
merged_df.info()
```

```python
# F2, F3, F4 Z-score의 가중평균

merged_df['Z_wt_sum'] = 0.5*merged_df['F2_Z'] + 0.3*merged_df['F3_Z'] + 0.2*merged_df['F4_Z']
merged_df
```

```python
# Z_wt_sum의 합 중 NA는 섹터평균으로 치환

merged_df['Z_wt_sum_NA_sct_mean'] = merged_df.groupby(['Sector'])['Z_wt_sum'].transform(lambda x: x.fillna(x.mean()))
merged_df
```

```python
# F1(시총) 랭크

merged_df['F1_pct_rank'] = merged_df['F1'].rank(ascending = False, pct = True)
merged_df
```

```python
# F1(시총) 랭크 상위 50%는 Z-score 가중평균, 하위 50%는 단순평균

import numpy as np

merged_df['Z_comb'] = np.where(merged_df['F1_pct_rank'] <= 0.5, merged_df['Z_wt_sum'], merged_df['Z_sum'])
merged_df
```

```python
# F2의 유니버스, 국가, 섹터별 Z-score 상관관계

merged_df[['F2','F2_Z','F2_ctr_Z','F2_sct_Z','Z_sum','Z_wt_sum']].corr()
```
